```python
import matplotlib.pyplot as plt
import numpy as np

# Data for performance vs cost analysis
systems_data = {
    'Gemini Assistant': {'performance': 16.20, 'cost': 0.004692},
    'PaperQA2': {'performance': 81.90, 'cost': 0.048798},
    'Modified PaperQA2': {'performance': 73.33, 'cost': 0.048798},
    'OpenAI Assistant': {'performance': 89.52, 'cost': 0.048798},
    'HybridOAIGem Assistant': {'performance': 85.71, 'cost': 0.003182},
    'HybridGemGem Assistant': {'performance': 84.76, 'cost': 0.003806},
    'VertexAI Assistant': {'performance': 86.67, 'cost': 0.000357},
    'Perplexity Assistant': {'performance': 17.14, 'cost': 0.0052}
}

# Extract data
systems = list(systems_data.keys())
costs = [systems_data[s]['cost'] * 100 for s in systems]  # Convert to cents
performance = [systems_data[s]['performance'] for s in systems]
efficiency = [p / (c / 100 * 1000) for p, c in zip(performance, costs)]  # Performance per $1000
short_names = [s.replace(' Assistant', '') for s in systems]

"""Radar chart showing performance, cost efficiency, and value"""
from math import pi

# Normalize metrics to 0-100 scale
norm_perf = performance
norm_cost_eff = [(100 - c/max(costs)*100) for c in costs]  # Inverted cost (lower is better)
norm_efficiency = [e/max(efficiency)*100 for e in efficiency]

categories = ['Performance', 'Cost Efficiency', 'Value Score']

fig, axes = plt.subplots(2, 4, figsize=(25, 10), subplot_kw=dict(projection='polar'))
axes = axes.flatten()

for i, system in enumerate(short_names):
    values = [norm_perf[i], norm_cost_eff[i], norm_efficiency[i]]
    values += values[:1]  # Complete the circle
    
    angles = [n / 3 * 2 * pi for n in range(3)]
    angles += angles[:1]
    
    axes[i].plot(angles, values, 'o-', linewidth=2)
    axes[i].fill(angles, values, alpha=0.25)
    axes[i].set_xticks(angles[:-1])
    
    # Custom positioning for labels
    label_angles = angles[:-1]
    for j, (angle, label) in enumerate(zip(label_angles, categories)):
        if label == 'Performance':
            # Move Performance label further to the right - ADJUST THIS VALUE
            axes[i].text(angle, 110, label, fontsize=14, fontweight='bold', 
                        ha='left', va='center', rotation=0)
        else:
            # Keep other labels in default position
            axes[i].text(angle, 120, label, fontsize=14, fontweight='bold', 
                        ha='center', va='center', rotation=0)
    
    axes[i].set_ylim(0, 100)
    axes[i].set_title(system, size=12, fontweight='bold', pad=20)
    axes[i].grid(True)
    axes[i].set_xticklabels([])  # Remove default labels since we're using custom ones

plt.tight_layout()
plt.savefig('../results/cost_efficiency.pdf')
```
